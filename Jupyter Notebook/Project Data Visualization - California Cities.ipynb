```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.available
plt.style.use("seaborn-whitegrid")
```

```python
cities = pd.read_csv("california_cities.csv")
cities.head()
```

```python
# Extract Latitude (Vi do) and Longitude (Kinh do)
lat, long = cities['latd'], cities['longd']
# Extract Population and Area
population, area = cities['population_total'], cities['area_total_km2']
```

```python
# Plot using Pyplot API
plt.figure(figsize = (8, 6)) # Set size of graph
plt.style.use("seaborn") # Set the style

plt.scatter(long, lat, s = area, c = np.log10(population), cmap = 'viridis', linewidths = 0, alpha = 0.7) # Plot the data

plt.colorbar(label = "log_${10}$(population)") # set colorbar

plt.axis("equal"); # balanced figure between horizontal and vertical 

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.clim(3, 7) # colorbar limit set

# Create a legend for cities' sizes
area_range = [50,100,300,500]
for area in area_range:
    plt.scatter([], [], s=area, c='k', alpha = 0.4, label = str(area) + " km$^2$")
plt.legend(labelspacing = 1, title = "Cities Area")

plt.title("California Cities: Area and Population Distribution");
```
