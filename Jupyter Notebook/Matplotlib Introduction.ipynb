### To draw plot -> 2 ways:
* Pyplot -> easy, fast
* Object Oriented API (OO) -> more advanced

#### Pyplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.available
plt.style.use("seaborn-whitegrid")

x = np.linspace(0, 10, 1000)
x

plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label="sin(x)")
plt.plot(x, np.cos(x), color = "red", label= "cos(x)")

plt.xlabel("Bien X")
plt.ylabel("Sin(X)")

#plt.xlim([0, 10])
#plt.ylim([-1, 1])
# <-> plt.axis
plt.axis([0, 4, -0.7, 0.6])
plt.axis("equal")

plt.title("Sin(X) and Cos(X)")

plt.legend(); # require label in plot function
```

#### Object-Oriented API

```python
# Step 1: Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Step 2: Set up plot
fig, ax = plt.subplots(figsize=(5,5))

# Step 3: Plot the data
ax.plot(x, y)
ax.set(title="A simple plot", xlabel = "x-axis", ylabel = "y-axis");
```

### Line chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create an array:
plt.style.use("seaborn-whitegrid")
x = np.linspace(0, 10, 100)
x[:5]
```

```python
# OO API plot a line chart
fig, ax = plt.subplots()
ax.plot(x, x**3);
```

### Scatter (Bieu do phan tan)

```python
# Pyplot API
plt.scatter(x, x**3);
```

```python
# OO API
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = rng.rand(100) * 1000

fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha = 0.3);
fig.colorbar(img1);
```

### Bar Chart

```python
# Prepare a Python dictionary
soft_drink_prices = {"Coca-Cola": 6, "Pepsi": 8, "Mirinda": 7}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title = "Soft Drink Prices", xlabel = "Soft Drink Type", ylabel = "Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
ax.set(title = "Soft Drink Prices", xlabel = "Price ($)", ylabel = "Soft Drink Type");
```

### Histogram (Bieu thi tan so dang cot)

```python
# Pyplot API (API = application programming interface -> giao dien lap trinh ung dung)
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
student_height[:10]
```

```python
plt.hist(student_height, bins = 50);
```

## Subplots (OO)

```python
# Option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (10, 5))

# Plot data to each axis
ax1.plot(x, x*2);
ax1.plot(x, x/2);

ax2.scatter(np.random.random(10), np.random.random(10));

ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax4.hist(student_height, bins = 50);
```

```python
# Option 2: See ax as an array
fig, ax = plt.subplots(2, 2, figsize = (10, 5))

# Plot data to each axis
ax[0, 0].plot(x, x*2);
ax[0, 0].plot(x, x/2);

ax[0, 1].scatter(np.random.random(10), np.random.random(10));

ax[1, 0].bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax[1, 1].hist(student_height, bins = 50);
```
