## Project: Nutrition Data Visualization on 80 cereal products

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Datasets
cereal = pd.read_csv("cereal.csv")
cereal.head()
```

```python
# Remove irrelevant fields 
fields = ["shelf", "weight", "cups", "rating"]
cereal_df = cereal.drop(fields, axis = 1)
```

```python
cereal_df
```

```python
# Correlation function -> corr()
cereal_corr = cereal_df.corr() # Get correlation data (only numerical data)
cereal_corr
```

```python
# ones_like can build a matrix of booleans (True, False) with the same shape as data
ones_corr = np.ones_like(cereal_corr, dtype = bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
# np's triu -> get upper triangle of the matrix
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data = cereal_corr, mask = mask);
sns.set_theme() # Set default theme for seaborn
```

```python
# When removing the upper triangle, here are still two empty cells in our matrix (calories and vitamins)
```

```python
adjusted_mask = mask[1:, :-1]
adjusted_mask
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
adjusted_cereal_corr 
```

```python
fig, ax = plt.subplots(figsize = (10, 8))

# That method use HuSL colors, so you need hue, saturation, and lightness
# I use hsluv.org to select color of this chart

cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap = True)

sns.heatmap(data = adjusted_cereal_corr, mask = adjusted_mask, 
            annot = True, annot_kws = {"fontsize": 13}, fmt = ".2f", cmap = cmap,
            vmin = -1, vmax = 1,
            linecolor = "white", linewidth = 1)

# Uppercase the label
yticks = [i.upper() for i in adjusted_cereal_corr.index]
xticks = [i.upper() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation = 0, fontsize = 13)
ax.set_xticklabels(xticks, rotation = 90, fontsize = 13);

# Making a title and adjust the font

title = "CORRELATION MATRIX\nSAMPLE CEREALS COMPOSITION\n"
ax.set_title(title, loc = "left", fontsize = 18);
```
