```python
cartier_ds['ref'] = cartier_ds['ref'].astype('string')
cartier_ds['categorie'] = cartier_ds['categorie'].astype('string')
cartier_ds['title'] = cartier_ds['title'].astype('string')
cartier_ds['tags'] = cartier_ds['tags'].astype('string')
cartier_ds['description'] = cartier_ds['description'].astype('string')
cartier_ds['image'] = cartier_ds['image'].astype('string')
```

```python
categories = cartier_ds['categorie'].unique()
categories
```

```python
categories_df = pd.DataFrame(data={'name': categories})
categories_df.head(5)
```

```python
categories_df['id'] = categories_df.index + 1
categories_df
```

```python
category_description = [
    'Rings are timeless pieces of jewelry that symbolize elegance and personal style. Whether worn as a statement accessory or a sentimental piece, they add a touch of sophistication to any outfit. From delicate bands to intricate designs, rings come in a variety of styles to suit every taste and occasion.'
    ,'Bracelets are versatile accessories that can effortlessly enhance any look, from casual to formal. They can be simple and understated or bold and eye-catching, often featuring unique charms, stones, or engravings. Perfect for layering or wearing alone, bracelets offer a delightful way to express individuality and fashion sense.',
    'Necklaces are captivating adornments that draw attention and add flair to the neckline. Available in an array of lengths and styles, they can range from minimalist pendants to elaborate statement pieces. Necklaces provide a perfect way to highlight one\'s personality and complement various outfits, making them a must-have in any jewelry collection.',
    'Earrings are essential accessories that can transform a look with their diverse styles and designs. From subtle studs to dramatic dangles, earrings cater to different preferences and occasions, offering a versatile way to accentuate facial features. They are an excellent way to express personal style and add a finishing touch to any ensemble.'
]

categories_df['description'] = category_description
categories_df.head(5)
```

```python
cartier_ds = cartier_ds.rename(columns={"ref": "code"})
```

```python
cartier_ds = cartier_ds.merge(categories_df[['id', 'name']], left_on='categorie', right_on='name')
```

```python
cartier_ds.drop(columns=['categorie'], inplace=True)
```

```python
cartier_ds
```

```python
cartier_ds = cartier_ds.rename(columns={"id": "category_id"})
```

```python
import random

random_rating_count = [random.randint(1, 100) for _ in range(len(cartier_ds))]
random_rating_avg = np.random.uniform(low=0.0, high=4.99, size=len(cartier_ds))
```

```python
cartier_ds['rating_count'] = random_rating_count
cartier_ds['average_rating'] = random_rating_avg
```

```python
cartier_ds
```

```python
public class Product: IEntity
{
    public Guid Id { get; set; }
    public string? Name { get; set; }
    public string Description { get; set; }
    public string? Code { get; set; }
#     public string? InternationalCode { get; set; } 
    public Guid CatalogId { get; set; }
    public float? AverageRating { get; set; }
    public long? RatingCount { get; set; }
    public DateTime Created_at { get; set; }
    public DateTime Updated_at { get; set; }
}
  public class ProductItem: IEntity
  {
      public Guid Id { get; set; }
      public string Size { get; set; }
      public string ProductSlug { get; set; }
      public string SKU { get; set; }
      public decimal UnitPrice { get; set; }
      public decimal DiscountPrice { get; set; }
      public int DiscountPercent { get; set; }    
      public int Stock { get; set;}
      public int State { get; set; }
  }
```
